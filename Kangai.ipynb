```python
import  pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
traffic =pd.read_csv("traffic.csv.zip")
traffic.head()
```

```python
traffic.info()
```

```python
traffic.tail()
```

```python
traffic
```

```python
sns.barplot(x=traffic['Junction'], y=traffic['Vehicles'], color='yellow')
```

```python
kangai=traffic.select_dtypes(include=['int', 'float'])

corr_matrix =kangai.corr()
plt.figure(figsize=(6, 3))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
```

```python
sns.boxplot(x='Junction', y='Vehicles', data=traffic, color='lightgreen')
plt.title('Box plot of Vehicles against Junction', color='Tomato')
```

```python
sns.histplot(x=traffic['Junction'], y=traffic['Vehicles'], data=traffic, color='violet')
plt.title('Histogram of Vehicles against Junction', color='red')
```

```python
sns.lineplot(x=traffic['Junction'],y=traffic['Vehicles'], data=traffic, color='black')
plt.title('A line graph of Junction against Vehicles')
```
