```python
a = int(input("enter first number: "))
b = int(input("Enter the second number: "))
def addition(a, b):
  return a + b
add = addition(a,b)
```

```python
def subtraction(a,b):
  return a - b

sub = subtraction(a, b)
```

```python
def multiplication(a, b):
  return a * b

mul = multiplication(a ,b)
```

```python
def division(a, b):
  return a // b

div = division(a, b)
```
